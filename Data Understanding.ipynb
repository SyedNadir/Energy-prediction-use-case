```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# graphics in svg format are more sharp and legible
%config InlineBackend.figure_format = 'svg'
```

# Data Understanding

```python
# Read the data
train = pd.read_csv('training.csv')
test = pd.read_csv('testing.csv')
```

```python
train.shape
```

```python
test.shape
```

## Exploratory data analysis

```python
# Merge data into 1 to perform Exploratory data analysis
energy = pd.concat([train, test])
```

```python
energy.shape
```

```python
energy.info()
```

```python
energy.describe().T
```

```python
# lights have many zeros
energy['lights'].value_counts().head()
```

```python
# Appliances have many outliers
energy['Appliances'].value_counts().tail(20)
```

```python
# correlation matrix
pd.options.display.float_format = '{:,.4f}'.format
energy.corr()
```

## Data Visualization

```python
# Box plots of lights energy utilization
sns.boxplot(x='lights', data=energy, orient = 'v')
```

```python
# Box plots of Appliances energy utilization
sns.boxplot(x='Appliances', data=energy, orient = 'v')
```

```python
energy.hist(figsize=(14,10))
```

```python
energy['Appliances'].hist(bins=50, figsize=(10,6))
```

```python
#Dist plot of Appliances
plt.rcParams['figure.figsize'] = 10,6
sns.distplot(energy['Appliances'])
```

```python
#Dist plot of lights
plt.rcParams['figure.figsize'] = 10,6
sns.distplot(energy['lights'])
```

```python
# Visualising the appliances, lights fixture energy usage
plt.plot(energy['Appliances'].to_numpy(), color = 'red', label = 'Appliances')
plt.plot(energy['lights'].to_numpy(), color = 'blue', label = 'lights')
plt.title('Appliances/lights Energy usage')
plt.xlabel('Time')
plt.ylabel('Energy usage')
plt.legend()
plt.show()
```

```python
# scatter plot matrix
%matplotlib inline
plt.style.use('ggplot')

from pandas.plotting import scatter_matrix

scatter_matrix(energy, figsize=(14,14))
plt.show()
```
